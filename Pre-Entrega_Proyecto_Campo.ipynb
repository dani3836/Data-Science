###**Definición de Objetivo**

Este estudio tiene como objetivo analizar los datos de obesidad recopilados en bases de datos públicas para impulsar el desarrollo y la venta de una aplicación de salud para dispositivos móviles que facilite a las personas lograr una pérdida de peso saludable.

```python
# Función para detectar atípicos

def detect_outliers_dataframe(df, interquartile_multiplier_threshold=1.5):
    for column in df.columns:
        if df[column].dtype == 'object':
            print(f"Skipping column '{column}' (categorical data)")
            print("===============================================================================")
            continue

        # Exclude binary columns
        unique_values = df[column].unique()
        if len(unique_values) == 2 and 0 in unique_values and 1 in unique_values:
            print(f"Skipping column '{column}' (binary data)")
            print("===============================================================================")
            continue

        data = df[column]
        q1 = np.percentile(data, 25)
        q3 = np.percentile(data, 75)
        iqr = q3 - q1

        lower_bound = q1 - interquartile_multiplier_threshold * iqr
        upper_bound = q3 + interquartile_multiplier_threshold * iqr

        outliers = [x for x in data if x < lower_bound or x > upper_bound]
        unique_outliers = list(set(outliers))  # Keep only unique outlier values

        if unique_outliers:
            print(f"🔥 Outliers in '{column}': {unique_outliers}")
            print(f"Number of rows with outliers in '{column}': {len(df[df[column].isin(outliers)])}")
        else:
            print(f"No outliers found in '{column}'")

        print("===============================================================================")
```
